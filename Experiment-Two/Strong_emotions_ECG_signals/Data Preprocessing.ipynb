# Libraries

```python
import pandas as pd
from datetime import datetime, timezone, timedelta

pd.options.display.max_columns = None
pd.options.display.max_colwidth = 9999
```

# Read ECG signals and Groundtruth/Label

```python
def read_ecgsignals(filename):
    '''
    Function takes file name with CSV extension constining ecg waveform signals

    Returns dataframe with ecgsignals  and timestamp with other variables
    '''
    
    #Read the csv file and harmonize datetime
    df = pd.read_csv(filename)
    df.reading_time = pd.to_datetime(df.reading_time)
    df.reading_time = df.reading_time + timedelta(hours=1)
    df.drop([23,42,43,44,45], inplace=True) #drop empty rows
    df_ecg  = df.copy()

    #rename timestamp to datetime and sort datetime
    df_ecg.rename(columns={'reading_time':'datetime'}, inplace=True)
    df_ecg.sort_values(by='datetime', ascending=True, inplace=True)
    df_ecg.reset_index(inplace=True)
    df_ecg.drop(['index'], axis=1, inplace=True)

    return df_ecg
```

```python
def read_label(filename):
    '''
    Function takes filename of groundtruth obatined from data annotation form

    Returns dataframe of label/groundtruth
    
    '''

    #read label
    df = pd.read_excel(filename)
    rename_dict = {'Endtime':'datetime', 'Soundgroup':'emotion', 'SoundID':'soundid'}
    drop_list = ['Starttime','Emotion','Emotionrating','UserID','Soundgroupselection']

    #drop selected columns, rename reatined columns and sort datetime
    df.drop(drop_list, axis=1, inplace=True)
    df.rename(columns=rename_dict, inplace=True)
    df.datetime = pd.to_datetime(df.datetime)
    df.drop([0], inplace=True)
    df.reset_index(inplace=True)
    df.drop(['index'], axis=1, inplace=True)
    df.sort_values(by='datetime',ascending=True,inplace=True)

    return df
```

# Merge ECG signals with label/ground-truth

```python
df_ecg = read_ecgsignals('MyFitbitData/Participant-One/Heart/Afib ECG Readings.csv')
df_label = read_label('Data-Annotation-Form-2/Excel Database 2.xlsx')

df_dataset = pd.merge(df_ecg, df_label, left_index=True, right_index=True, how='outer')
df_dataset.to_csv('df_dataset.csv')
print('saved')
```
